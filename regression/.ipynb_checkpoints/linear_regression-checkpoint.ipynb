```python
import numpy as np
import matplotlib
matplotlib.use('agg')
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
N_train = 400
N_test = 100
noise = 50000
X_train = np.random.randint(low=200,high=2000,size=N_train)
X_test = np.random.randint(low=200,high=2000,size=N_test)

Y_train = 500*X_train + noise*np.random.randn(N_train)*1.5*X_train/np.mean(X_train)+5e4
Y_test = 500*X_test + 1e-2*X_test**2 + noise*np.random.randn(N_test)*1.5*X_test/np.mean(X_test)+5e4
```

```python
plt.scatter(X_train,Y_train, color='r', label='Training Data')
plt.xlabel('Square feet')
plt.ylabel('House price ($)')
plt.legend(loc='upper left')
plt.show()
```
